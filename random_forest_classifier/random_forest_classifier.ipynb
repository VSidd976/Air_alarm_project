```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.utils import resample
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from joblib import dump, load
```

# Data processing

```python
df = pd.read_csv("df+dist+hol.csv")
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df
```

```python
df.isna().sum()
```

```python
df['text_vector'] = df['text_vector'].fillna(0)
```

```python
df.isna().sum()
```

```python
df['alarms'].unique()
```

```python
df['alarms'] = df['alarms'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df['alarms'].unique()
```

```python
df.columns
```

```python
df = df.drop(columns=['anomaly', 'date'])
```

```python
df = df.sort_values(by=['city_address', 'datetime']).reset_index(drop=True)
```

```python
df
```

```python
df = df.drop(columns=['city_address'])
```

```python
exclude_cols = ['alarms', 'datetime']
feature_cols = [col for col in df.columns if col not in exclude_cols]
```

```python
scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])
```

# Oversampling data

```python
df_majority = df[df['alarms'] == 0]
df_minority = df[df['alarms'] == 1]
```

```python
df_minority_upsampled = resample(
    df_minority,
    replace=True,
    n_samples=len(df_majority),
    random_state=42
)
```

```python
df_balanced = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_balanced['alarms'].value_counts()
```

```python
df_balanced
```

```python
X = df_balanced[feature_cols]
y = df_balanced['alarms']
```

```python
tscv = TimeSeriesSplit(n_splits=5)

for train_index, val_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[val_index]
    y_train, y_test = y.iloc[train_index], y.iloc[val_index]
```

# Parameters searching

```python
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}
```

```python
model = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)
```

```python
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
X_sample, y_sample = resample(X, y, n_samples=100_000, random_state=42)
```

```python
X_sample.isna().sum()
```

```python
y_sample.isna().sum()
```

```python
random_search.fit(X_sample, y_sample)
```

```python
random_search.best_estimator_
```

```python
random_search.best_estimator_.get_params()
```

# Model training

```python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,
    min_samples_split=5,
    min_samples_leaf=1,
    max_features=None,
    bootstrap=True,
    n_jobs=-1,
    random_state=42,
    class_weight='balanced'
)
model.fit(X_train, y_train)
```

# Evaluating model

```python
model = load("random_forest_classifier.pkl")
```

```python
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Prediction')
plt.ylabel('True')
plt.title('Confusion matrix')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='b', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

# Top 20 features

```python
importances = model.feature_importances_
```

```python
features = pd.DataFrame({"Feature": feature_cols, "Importance": importances})
```

```python
sorted_features = features.sort_values(by="Importance", ascending=False).head(20)
```

```python
plt.figure(figsize=(12, 8))

norm = plt.Normalize(sorted_features['Importance'].min(), sorted_features['Importance'].max())
sm = plt.cm.ScalarMappable(cmap="viridis_r", norm=norm)
sm.set_array([])

ax = sns.barplot(
    x="Importance",
    y="Feature",
    data=sorted_features,
    hue="Feature",
    palette=sns.color_palette("viridis", n_colors=len(sorted_features)),
    dodge=False,
    legend=False
)

cbar = plt.colorbar(sm, ax=ax)
cbar.set_label('Feature Importance')

plt.title("Top 20 the most important features")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
