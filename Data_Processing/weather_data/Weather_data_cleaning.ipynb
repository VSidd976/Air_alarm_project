hour_solarradiation, hour_uvindex, day_solarenergy, day_solarradiation, day_uvindex, hour_precip - have pretty small percentage of missing values, that could be replaced. hour_preciptype and hour_preciptype has to many missing values and provides similar type of information to day_conditions column - comsidering all of that we will be dropping it. Column hour_visibility has 45% missing value, but may be extremly important for identifying suitable conditions for air-attack, so we will try to fill the values with present values of day_visibility and wil the others as weighted mean of the closest available. Same we will do for hour_solarenergy.

```python
def fill_hour_with_daily(row, col):
    if not pd.isnull(row[col]):
        return row[col]
    elif not pd.isnull(row[col]):
        return row[col]
    else:
        return np.nan
```

```python
df['hour_visibility'] = df.apply(lambda row: fill_hour_with_daily(row, 'hour_visibility'), axis=1)
```

```python
df['hour_visibility'] = df['hour_visibility'].interpolate(method='linear', limit_direction='both')
```

```python
df['hour_solarenergy'] = df.apply(lambda row: fill_hour_with_daily(row, 'hour_solarenergy'), axis=1)
```

```python
df['hour_solarenergy'] = df['hour_solarenergy'].interpolate(method='linear', limit_direction='both')
```

```python
cols_to_fill = ["hour_solarradiation", "hour_uvindex", "day_solarenergy", "day_solarradiation", "day_uvindex", "hour_precip", "day_visibility"]
```

```python
for col in cols_to_fill:
    df[col] = df[col].interpolate(method='linear', limit_direction='both')
```

```python
df = df.drop(columns=['day_preciptype'])
```

```python
df = df.drop(columns=['hour_preciptype'])
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
corr_pairs = corr_matrix.unstack()
```

```python
corr_pairs = corr_pairs[corr_pairs < 1]
```

```python
corr_pairs = corr_pairs.drop_duplicates()
```

```python
threshold = 0.7
strong_corrs = corr_pairs[corr_pairs > threshold]
strong_corrs
```

# Outliers

```python
mean_temp = df['hour_temp'].mean()
std_temp = df['hour_temp'].std()
threshold_upper = mean_temp + 2 * std_temp
threshold_lower = mean_temp - 2 * std_temp

df['anomaly'] = (df['hour_temp'] > threshold_upper) | (df['hour_temp'] < threshold_lower)

plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='datetime', y='hour_temp', label='Temperature')
plt.scatter(df[df['anomaly']]['datetime'], df[df['anomaly']]['hour_temp'], color='red', label='Anomalies', zorder=5)
plt.title('Temperature anomalies')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
