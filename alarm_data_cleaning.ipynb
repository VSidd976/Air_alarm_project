#**Find and remove duplicates**

```python
duplicates = df[df.duplicated(subset=['region_city', 'start'], keep=False)]

if not duplicates.empty:
    print("Duplicates found:")
    print(duplicates)

    df = df.drop_duplicates(subset=['region_city', 'start'], keep='first')
    print("\nDuplicates removed.")
else:
    print("No duplicates found.")

duplicates_after = df[df.duplicated(subset=['region_city', 'start'], keep=False)]

if not duplicates_after.empty:
    print("Duplicates found after removal:")
    print(duplicates_after)
else:
    print("No duplicates found after deletion.")
```

##**Creating new columns to better train the model**

```python
df['duration_hours'] = (df['end'] - df['start']).dt.total_seconds() / 3600

df['day_of_week'] = df['start'].dt.day_name()

df['is_weekend'] = df['start'].dt.dayofweek >= 5
df['is_weekend'] = df['is_weekend'].astype(int)
```

```python
events = pd.concat([
    df[['start']].rename(columns={'start': 'time'}).assign(change=1),
    df[['end']].rename(columns={'end': 'time'}).assign(change=-1)
])

events = events.sort_values('time')
events['active_regions_count'] = events['change'].cumsum()

df = pd.merge_asof(
    df.sort_values('start'),
    events[['time', 'active_regions_count']].sort_values('time'),
    left_on='start',
    right_on='time'
)
df = df.drop(columns='time')
```

### **Removing the 'start', 'end' columns, as we no longer need them after dividing them into separate columns**

```python
df = df.drop(columns=['start', 'end'])
```

### **Allocation of areas to each separate column**

```python
dummies = pd.get_dummies(df['region_city'], drop_first=True)
dummies = dummies.astype(int)
df = df.drop(columns=['region_city']).join(dummies)
df
```
