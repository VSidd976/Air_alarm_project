```python
month_map = {
    'січня': '01', 'лютого': '02', 'березня': '03', 'квітня': '04',
    'травня': '05', 'червня': '06', 'липня': '07', 'серпня': '08',
    'вересня': '09', 'жовтня': '10', 'листопада': '11', 'грудня': '12'
}

def convert_to_date(date_str, year):
    parts = date_str.split()
    if len(parts) == 2 and parts[1] in month_map:
        day = parts[0].zfill(2)
        month = month_map[parts[1]]
        return pd.to_datetime(f'{year}-{month}-{day}', format='%Y-%m-%d')
    return pd.NaT

years = [2022, 2023, 2024, 2025]
dfh['key'] = 1  

year_df = pd.DataFrame({'year': years})
year_df['key'] = 1

expanded_df = dfh.merge(year_df, on='key').drop(columns=['key'])

expanded_df['date_converted'] = expanded_df.apply(
    lambda row: convert_to_date(row['date'], row['year']) if 'date' in row else pd.NaT,
    axis=1
)
```

```python
from datetime import datetime
import calendar

weekdays = {
    'понеділок': 0,
    'вівторок': 1,
    'середа': 2,
    'четвер': 3,
    'пʼятниця': 4,
    'субота': 5,
    'неділя': 6,
}

ordinals = {
    'перша': 1,
    'друга': 2,
    'третя': 3,
    'четверта': 4,
    'остання': -1,
}

months = {
    'січня': 1, 'лютого': 2, 'березня': 3, 'квітня': 4, 'травня': 5,
    'червня': 6, 'липня': 7, 'серпня': 8, 'вересня': 9, 'жовтня': 10,
    'листопада': 11, 'грудня': 12
}

def get_relative_date(year, phrase):
    words = phrase.lower().split()
    if len(words) < 3:
        return None

    try:
        ordinal = ordinals.get(words[0])
        weekday = weekdays.get(words[1])
        month = months.get(words[-1])
        
        if ordinal is None or weekday is None or month is None:
            return None

        cal = calendar.Calendar()
        days = [d for d in cal.itermonthdates(year, month) if d.weekday() == weekday and d.month == month]

        if ordinal > 0:
            return days[ordinal - 1]
        elif ordinal == -1:
            return days[-1]
    except Exception as e:
        return None

years = [2022, 2023, 2024, 2025]
phrases = ['Друга субота березня', 'Третя неділя березня',
       'Перша неділя квітня', 'Третя субота квітня',
       'Друга неділя травня', 'третій четвер травня',
       'третя субота травня', 'третя неділя травня',
       'остання субота травня', 'остання неділя травня',
       'перша неділя червня',
       'друга неділя червня', 'третя неділя червня',
       'остання неділя червня', 'перша неділя липня',
       'друга неділя липня', 'третя неділя липня', 'Остання неділя липня',
       'перша неділя серпня', 'друга неділя серпня',
       'остання субота серпня', 'остання неділя серпня',
       'перша неділя вересня', 'друга субота вересня',
       'друга неділя вересня', 'третя субота вересня',
       'четверта неділя вересня', 'перша неділя жовтня',
       'друга неділя жовтня', 'третя субота жовтня',
       'третя неділя жовтня', 'остання неділя жовтня',
       'перша неділя листопада', 'третя неділя листопада',
       'четверта субота листопада', 'друга неділя грудня']

results = []
for phrase in phrases:
    for y in years:
        date = get_relative_date(y, phrase)
        results.append({'holiday': phrase, 'year': y, 'date': date})

df_result = pd.DataFrame(results)
print(df_result)
```
