```python
import pandas as pd

# Ensure datetime columns are in datetime format
dfa['start'] = pd.to_datetime(dfa['start'])
dfa['end'] = pd.to_datetime(dfa['end'])
dfw['datetime'] = pd.to_datetime(dfw['datetime'])

# Mapping dictionary
ukr_to_eng_oblasts = {
    'Львівська обл.': 'Lviv',
    'Чернігівська обл.': 'Chernihiv',
    'Вінницька обл.': 'Vinnytsia',
    'Харківська обл.': 'Kharkiv',
    'Тернопільська обл.': 'Ternopil',
    'Одеська обл.': 'Odesa',
    'Дніпропетровська обл.': 'Dnipro',
    'Запорізька обл.': 'Zaporozhye',
    'Полтавська обл.': 'Poltava',
    'Сумська обл.': 'Sumy',
    'Житомирська обл.': 'Zhytomyr',
    'Київська обл.': 'Kyiv',
    'Черкаська обл.': 'Cherkasy',
    'Кіровоградська обл.': 'Kropyvnytskyi',
    'Миколаївська обл.': 'Mykolaiv',
    'Хмельницька обл.': 'Khmelnytskyi',
    'Рівненська обл.': 'Rivne',
    'Івано-Франківська обл.': 'Ivano-Frankivsk',
    'Закарпатська обл.': 'Uzhgorod',
    'Луганська обл.': 'Luhansk',
    'Донецька обл.': 'Donetsk',
    'АР Крим': 'Simferopol',
    'Севастополь': 'Simferopol', 
    'Київ': 'Kyiv'
}

# Translate region_city to match city_address
dfa['region_city_eng'] = dfa['region_city'].map(ukr_to_eng_oblasts)

# Initialize 'alarms' column
dfw['alarms'] = 0

# Function to count overlapping alerts
def count_alarms(row):
    city = row['city_address']
    t_start = row['datetime']
    t_end = t_start + pd.Timedelta(hours=1)
    
    mask = (
        (dfa['region_city_eng'] == city) &
        (dfa['start'] <= t_end) &
        (dfa['end'] >= t_start)
    )
    return mask.sum()

# Optional: show progress bar
from tqdm import tqdm
tqdm.pandas()

# Apply function
dfw['alarms'] = dfw.progress_apply(count_alarms, axis=1)
```
